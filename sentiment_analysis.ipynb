```python
def clean_text(doc):
    doc = doc.lower()
    doc=emoji.demojize(doc).replace(':','')
    doc = re.sub(r'[^A-z0-9 ]+','',doc)
    doc_list = doc.split()
    doc_list = [word for word in doc_list if word not in sw]
    doc_list = [lemmatizer.lemmatize(i) for i in doc_list]
    return " ".join(doc_list)
```

```python
def polarity(n):
    return analyzer.polarity_scores(n)["compound"]

df["polarity"]=df["Text_cleaned"].apply(polarity)
```

```python
def sentiment(x):
    if x>0.5:
        return "positive"
    elif -0.3<x<0.5:
        return "neutral"
    else:
        return "negative"
df["sentiment"]=df["polarity"].apply(sentiment) 
```

```python
pipe = Pipeline(
    [
        ('vectorization', CountVectorizer()),
        ('nb', MultinomialNB())
    ]
)

MAX_FEATURES = [1000, 1500, 2000]
ALPHA = [1, 10]

# Observe the Key Value Pair format
parameter_grid = [{'vectorization__preprocessor' : [clean_text],
                   'vectorization__max_features' : MAX_FEATURES, 
                   'nb__alpha' : ALPHA}]

clf = GridSearchCV(
    estimator=pipe, 
    param_grid=parameter_grid, 
    scoring='accuracy',
    cv=5,
    return_train_score=True,
    verbose=1
)

%time clf.fit(X_train, Y_train)

print("Best estimator found on train set")
print(clf.best_estimator_)
print()

print('Score on Test Data: ', clf.score(X_test, Y_test))
```
